```python
def split_samples(filenames, img_type):
  """
  this function splits the patch images by samples according to their names
  """
    num = len(filenames)
    sample_name_id = []
    sample_full_name = []
    
    if img_type == 'msk':
        pos = -3
        posi = -9
    else:
        pos = -2
        posi = -4
        
    for i in range(num):
        curr_name = filenames[i].split('/')[-1]
        
        sample_id = curr_name.split('_')[pos]
#         full_name = curr_name[:posi]
        
      
            
        sample_name_id.append(sample_id)
#         sample_full_name.append(full_name)
        
    return sample_name_id
        
        
```

```python
def split_donors(filenames):
  """
  this function splits the patch images by donors according to their names
  """
    num = len(filenames)
    donor_name = []


        
    for i in range(num):
        curr_name = filenames[i].split('/')[-1]
        
        donors = curr_name.split('_')[0]

        
      
            
        donor_name.append(donors)

        
    return donor_name
        
        
```

```python
def copy_files(fnames, idx_list, dst):
  """
  this function copy and paste images to the new directory according to the input image filenames
  """
    for i in range(len(idx_list)):
        old_name = fnames[idx_list[i]].split('/')[-1]
        new_dst = dst + old_name
        shutil.copyfile(fnames[idx_list[i]], new_dst)
```

When preparing the dataset, in order to prevent data leakage, the train/val/test dataset should not contain patches/images from same donors. Based on our current dataset, 

*   H1 and LK75, LK63 are assigned to training dataset;
*   H2, LK67, and LK51 are assigned to validation dataset;
*   H3, LK2, and LK18 are assigned to test dataset. 

The number of patches selected from each donor were chose based on the minimum number of patches from the donors in each dataset, in order to balance data in the HC (0) and CLL (1) group.
